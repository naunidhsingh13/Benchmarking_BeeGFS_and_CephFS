```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from io import BytesIO
```

```python
file_path = "BENCHMARKS/csv_stuff-20210422234511/net.csv"
```

```python
content = ""
with open(file_path, 'r') as f:
    for line in f.readlines():
        if "#" in line:
            header = line
        else:
            content += line

content = header+content
```

```python
header
```

```python
content_stream = BytesIO(content.encode())
```

```python
df = pd.read_csv(content_stream, delimiter=';')
```

```python
grpd = df.groupby("# hostname")
```

```python
s1_df = grpd.get_group("an-sap-s-1").reset_index(drop=True)
```

```python
s2_df = grpd.get_group("an-sap-s-2").reset_index(drop=True)
```
